```python
# Loading necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

# Unzip Fleet.zip using your File Explorer
After that you should have a folder 'Fleet', which contains the database Fleet.db

```python
# Open our Fleet.db database file using the SQLite database engine
# Change the following to be the path of your Fleet.db file
import sqlite3
connection = sqlite3.connect("Fleet.db")
print("Database connected")
```

```python
# Execute a query, and return the result as a Pandas dataframe
# This database has one table called Fleet. This query gets the first 10 rows from Fleet
# Note: Not all columns are shown as we have so many!
dataframe = pd.read_sql_query('SELECT * FROM Fleet LIMIT 10', connection)
dataframe
```

```python
# Get the names of the columns using the SQLite "description".
connection.execute('select * from Fleet').description
```

```python
# Get the number of rows. This may take a while!
df = pd.read_sql_query('SELECT COUNT(*) FROM Fleet', connection)
df
```

```python
# How many Toyotas are there?
df = pd.read_sql_query('SELECT COUNT(*) FROM Fleet WHERE MAKE="TOYOTA"', connection)
df
```

```python
# Display the number of cars of each MAKE for the first 100 makes
df = pd.read_sql_query('SELECT MAKE, COUNT(MAKE) FROM Fleet GROUP BY MAKE LIMIT 100', connection)
df
```

```python
# Plot the number of cars of each MAKE for the first 20 makes
sql = """SELECT MAKE, COUNT(MAKE) AS NUMCARS 
            FROM Fleet GROUP BY MAKE LIMIT 20"""
df = pd.read_sql_query(sql, connection)
df.plot.bar(x='MAKE')
#trace = plotly.graph_objs.Bar(x=dataframe.MAKE,y=dataframe.NUMCARS)
plt.title("Number of Cars by Make")
plt.xlabel('Car Make')
plt.ylabel('Number of cars')
```

## 1. What is the earliest year of first registration for the cars?
(2 marks)

```python
sql = 'SELECT MIN(FIRST_NZ_REGISTRATION_YEAR) AS FIRST_REG_YEAR FROM Fleet'
pd.read_sql_query(sql, connection)
#1899
```

## 2.	What are the makes, models and vehicle year of all the cars with a vehicle year earlier than 1900?
(2 marks)

```python
sql = 'SELECT MAKE, MODEL, VEHICLE_YEAR FROM Fleet WHERE VEHICLE_YEAR<1900'
pd.read_sql_query(sql, connection)
# 15 results
```

## 3.	What are the 10 most popular (by count) car makes, and the counts of these? 
(2 marks)

```python
sql = """SELECT MAKE, COUNT(MAKE) AS NUMCARS FROM Fleet GROUP BY MAKE ORDER BY NUMCARS DESC LIMIT 10"""
pd.read_sql_query(sql, connection)
# Toyota most popular, then nissan, trailers, mazda, ford etc.
```

## 4.	What are the 20 most popular (by count) car ‘models’, and the counts of these? 
Note that we consider a ‘model’ to be defined by both the MAKE and MODEL fields, such as a ‘Holden’ ‘Commodore’. Therefore, your query needs to count each ‘make’, ‘model’ pair separately.
(3 marks)

```python
sql = """SELECT MAKE, MODEL, COUNT(MODEL) AS NUMCARS FROM Fleet GROUP BY MAKE, MODEL ORDER BY NUMCARS DESC LIMIT 20"""
pd.read_sql_query(sql, connection)
# Toyota Corolla wins this unsurprisingly
```

## 5. How many cars are first registered in each of the most recent 50 years? 
You may end up with a NULL year in your result, meaning only 49 years are actually shown; that is ok in this answer, and also in your later answers.
(3 marks)

```python
sql = """SELECT FIRST_NZ_REGISTRATION_YEAR, COUNT(FIRST_NZ_REGISTRATION_YEAR) AS NUM_CARS_REG FROM Fleet 
            GROUP BY FIRST_NZ_REGISTRATION_YEAR ORDER BY FIRST_NZ_REGISTRATION_YEAR DESC LIMIT 50"""
reg_by_year = pd.read_sql_query(sql, connection)
reg_by_year
# More recent years have more cars registered
```

## 6. Generate a labelled bar graph plot of the previous answer (with the dates in ascending order).
(3 marks)

```python
rev = reg_by_year.iloc[::-1]
# Reverse dataframe to plot ascending
rev.plot.bar(x='FIRST_NZ_REGISTRATION_YEAR', y='NUM_CARS_REG')
plt.title("Number of Cars First Registration by Year")
plt.xlabel('Year')
plt.ylabel('Number of cars Registered')
# Code from the web to clean up the x-labels by only displaying every 2nd year-label
ax = plt.gca()
temp = ax.xaxis.get_ticklabels()
temp = list(set(temp) - set(temp[::2]))
for label in temp:
    label.set_visible(False)
```

## 7. How many Toyota cars were first registered in each year from 1950 onwards? 
Hint: You can use AND to join together multiple tests in a WHERE.
(3 marks)

```python
sql = """SELECT FIRST_NZ_REGISTRATION_YEAR, COUNT(FIRST_NZ_REGISTRATION_YEAR) AS NUM_TOYOTA_REG FROM Fleet 
        WHERE MAKE='TOYOTA' AND FIRST_NZ_REGISTRATION_YEAR>=1950
            GROUP BY FIRST_NZ_REGISTRATION_YEAR ORDER BY FIRST_NZ_REGISTRATION_YEAR"""
toy_by_year = pd.read_sql_query(sql, connection)
toy_by_year
# First toyota in 1966
```

## 8.	How many cars from Japan (ORIGINAL_COUNTRY=”JAPAN”) were first registered in each year from 1950 onwards? 
(marks 3)

```python
sql = """SELECT FIRST_NZ_REGISTRATION_YEAR, COUNT(FIRST_NZ_REGISTRATION_YEAR) AS NUM_JAP_CARS_REG FROM Fleet 
        WHERE ORIGINAL_COUNTRY='JAPAN' AND FIRST_NZ_REGISTRATION_YEAR>=1950
            GROUP BY FIRST_NZ_REGISTRATION_YEAR ORDER BY FIRST_NZ_REGISTRATION_YEAR"""
jap_by_year = pd.read_sql_query(sql, connection)
jap_by_year = jap_by_year[:-1]  #Remove null row
jap_by_year
# Lots
```

## 9.	How many cars from Germany were first registered in each year from 1950 onwards? 
(3 marks)

```python
sql = """SELECT FIRST_NZ_REGISTRATION_YEAR, COUNT(FIRST_NZ_REGISTRATION_YEAR) AS NUM_GER_CARS_REG FROM Fleet 
        WHERE ORIGINAL_COUNTRY='GERMANY' AND FIRST_NZ_REGISTRATION_YEAR>=1950
            GROUP BY FIRST_NZ_REGISTRATION_YEAR ORDER BY FIRST_NZ_REGISTRATION_YEAR"""
ger_by_year = pd.read_sql_query(sql, connection)
ger_by_year = ger_by_year[:-1]  #Remove null row
ger_by_year
# Quite a lot less recently
```

## 10. Generate a labelled barplot (with a legend) showing all first-registered data for Japan and Germany. 
Hint: Combine the results from the previous two questions with `pd.merge()`.

(3 marks)

```python
joined = pd.merge(jap_by_year, ger_by_year, how='outer', on="FIRST_NZ_REGISTRATION_YEAR")
joined = joined.astype({'FIRST_NZ_REGISTRATION_YEAR':'int'})
#joined = joined['FIRST_NZ_REGISTRATION_YEAR'>2000]
joined.sort_values('FIRST_NZ_REGISTRATION_YEAR', inplace=True)

plt.rcParams["figure.figsize"] = (20,12)
joined.plot.bar(x='FIRST_NZ_REGISTRATION_YEAR', width=1)
plt.title("Number of Cars First Registration by Year")
plt.xlabel('Year')
plt.ylabel('Number of cars Registered')
#plt.yscale('log')                        # <- uncomment this if you want to see details on the early years

# Code from the web to clean up the x-labels by only displaying every 3rd year-label
ax = plt.gca()
temp = ax.xaxis.get_ticklabels()
temp = list(set(temp) - set(temp[::3]))
for label in temp:
    label.set_visible(False)
    
```
