```python
# Use Pandas to load a CSV file into a dataframe (i.e. a table)
import pandas
dataframe = pandas.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv")
# display the first few lines of the table
dataframe.head()
```

```python
# Use Pandas to create bar plot
dataframe.plot.bar(x='School', y='Gap', rot=90)
```

```python
# Use Pandas to create bar plot
# combine with matplotlib to modify labels
dataframe.plot.bar(x='School', y='Gap', rot=90)
plt.title("Gap in Average Earnings for Male and Female Graduates")
plt.ylabel('Salary Gap (in thousands)')
```

```python
#Create plot and export as PDF
dataframe.plot.bar(x='School')
plt.title("Gap in Average Earnings for Male and Female Graduates")
plt.ylabel('Salary Gap (in thousands)')
plt.savefig('salary_gap.pdf', bbox_inches='tight')
```
